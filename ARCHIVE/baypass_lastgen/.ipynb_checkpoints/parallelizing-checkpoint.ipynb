```python
# scontrol show node
nodes_caltech = 'nodes_caltech.txt'
```

```python
import pandas as pd

def parse_node_info(filepath):
    # Initialize an empty list to hold all node dictionaries
    nodes = []
    node = {}

    # Open and read the file
    with open(filepath, 'r') as file:
        for line in file:
            # Clean up extra spaces and split on single spaces
            parts = line.strip().split()
            # Iterate over parts and split key=value pairs
            for part in parts:
                if '=' in part:
                    key, value = part.split('=', 1)
                    node[key] = value

            # Check if we've reached the end of a node's information
            if line.strip() == '' and node:  # Assuming blank lines between nodes
                nodes.append(node)
                node = {}  # Reset for next node

        # Append the last node if not already appended
        if node:
            nodes.append(node)

    # Create a DataFrame from the list of dictionaries
    df = pd.DataFrame(nodes)
    return df

# Usage
node_df = parse_node_info(nodes_caltech)
```
