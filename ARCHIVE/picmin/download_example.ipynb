```python
import requests

# URLs for the CSV files
urls = [
    "https://raw.githubusercontent.com/TBooker/PicMin/main/vignettes/lineage_1.csv",
    "https://raw.githubusercontent.com/TBooker/PicMin/main/vignettes/lineage_2.csv",
    "https://raw.githubusercontent.com/TBooker/PicMin/main/vignettes/lineage_3.csv",
    "https://raw.githubusercontent.com/TBooker/PicMin/main/vignettes/lineage_4.csv",
    "https://raw.githubusercontent.com/TBooker/PicMin/main/vignettes/lineage_5.csv",
    "https://raw.githubusercontent.com/TBooker/PicMin/main/vignettes/lineage_6.csv",
    "https://raw.githubusercontent.com/TBooker/PicMin/main/vignettes/lineage_7.csv",
]

# Loop through each URL and download the file
for i, url in enumerate(urls, start=1):
    response = requests.get(url)
    # Save the content to a CSV file
    with open(f'lineage_{i}.csv', 'wb') as file:
        file.write(response.content)
    print(f'lineage_{i}.csv downloaded successfully.')
```

```python
# create sbatch files to submit on cedar server
shfiles = []

file = 'picmin.sh'
cmd = f'Rscript run_pic_min.r'
text = f'''#!/bin/bash
#SBATCH --job-name=picmin
#SBATCH --time=1-00:00:00 
#SBATCH --ntasks=1
#SBATCH --mem-per-cpu=30gb
#SBATCH --output=picmin_%j.out
#SBATCH --mail-user=tbellagio@carnegiescience.edu
#SBATCH --mail-type=FAIL

source /home/tbellagio/miniforge3/etc/profile.d/conda.sh

export PATH="${{PATH}}:/home/username/bin"

cd '/carnegie/nobackup/scratch/tbellagio/gea_grene-net/picmin'

conda activate /home/tbellagio/miniforge3/envs/r-environment
{cmd}


'''
with open(file, 'w') as o:
    o.write("%s" % text)
shfiles.append(file)
```

```python
## now run the shfiles
for shfile in shfiles:
    # Submit each sbatch script to the SLURM scheduler
    subprocess.run(["sbatch", shfile], check=True)
```
