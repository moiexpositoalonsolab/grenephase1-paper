```python
# Create the parent directory for splits
os.makedirs(parent_dir, exist_ok=True)

for idx, (train_samples, test_samples) in enumerate(splits_samples):
    # Create a dedicated folder for each split
    split_dir = os.path.join(parent_dir, f"split_{idx+1}")
    os.makedirs(split_dir, exist_ok=True)

    # Convert train and test samples to DataFrames
    train_df = pd.DataFrame(train_samples, columns=['sample'])
    test_df = pd.DataFrame(test_samples, columns=['sample'])

    # Save the train and test samples as CSV files
    train_df.to_csv(os.path.join(split_dir, "train_samples.csv"), index=False)
    test_df.to_csv(os.path.join(split_dir, "test_samples.csv"), index=False)
```
