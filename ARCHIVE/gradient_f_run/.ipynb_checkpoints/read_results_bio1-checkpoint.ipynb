```python
import pandas as pd
import os
```

```python


# Define the parent directory where all splits are located
parent_dir = "splits_samples/"

# Initialize an empty list to store the results
offset_list = []

# Walk through all directories and subdirectories
for root, dirs, files in os.walk(parent_dir):
    for file in files:
        # Check if the file matches the genomic offset pattern
        if file.startswith("offsets_split_") and file.endswith(".csv"):
            file_path = os.path.join(root, file)
            
            # Read the CSV file
            offset_data = pd.read_csv(file_path)
            
            # Extract split and batch numbers from the filename
            split_num = int(file.split("/")[-1].split("_")[2].replace('.csv', ''))
            site_train = pd.read_csv(f'splits_samples/split_{split_num}/train_samples.csv')['sample'].str.split('_').str[0].drop_duplicates().reset_index(drop=True)
            site_test = pd.read_csv(f'splits_samples/split_{split_num}/test_samples.csv')['sample'].str.split('_').str[0].drop_duplicates().reset_index(drop=True)
            
            # Add split and batch information to the DataFrame
            offset_data['split'] = split_num
            offset_data['site_train'] = site_train
            offset_data['site_test'] = site_test.to_list() * 30
            # Append the data to the list
            offset_list.append(offset_data)

# Combine all data frames into one
combined_offsets = pd.concat(offset_list, ignore_index=True)
```

```python
offset_gf = combined_offsets.groupby('site_test')['offsets'].mean().reset_index()
```

```python
offset_gf.to_csv('go_gf_l1o.csv',index=None)
```
