```python
path = '/carnegie/nobackup/scratch/tbellagio/gea_grene-net/jacknife_first_gen/'
```

```python
## create the env testgin 

with open('splits_samples_first_gen.pkl', 'rb') as file:
    # Load data from the file
    splits_samples = pickle.load(file)
```

```python
len(splits_samples)
```

```python
for i in range(len(splits_samples)):
## get the training and testing sites 
    train_samples = splits_samples[i][0]
    test_samples = splits_samples[i][1]
    
    clim_sites_during_exp = pd.read_csv('../key_files/bioclimvars_sites_era5_year_2018.csv')
    
    sites_af = pd.Series(train_samples).str.split('_').str[0].astype(int)
    
    sites_af.name = 'site'
    ## env trainign 
    env_train = sites_af.reset_index().merge(clim_sites_during_exp).drop(['index'],axis=1)

    ## for now only work with bio1 and bio17 
    env_train = env_train[['site', 'bio1', 'bio17']]
    
    mean_bio1 = np.mean(env_train['bio1'])
    mean_bio17 = np.mean(env_train['bio17'])
    std_bio1 = np.std(env_train['bio1'])
    std_bio17 = np.std(env_train['bio17'])

    ## standarize training dataset 
    env_train['bio1'] = (env_train['bio1'] - mean_bio1) / std_bio1 
    env_train['bio17'] = (env_train['bio17'] - mean_bio17) / std_bio17

    file_train = path + f'split_number_{i}/env_training_sites.csv'
    env_train.to_csv(file_train)
    
    ## for the testing
    
    sites_af = pd.Series(test_samples).str.split('_').str[0].astype(int)
    sites_af.name = 'site'
    env_test = sites_af.reset_index().merge(clim_sites_during_exp).drop(['index'],axis=1)
    env_test = env_test[['site', 'bio1', 'bio17']]

    ##scale it
    
    env_test['bio1'] = (env_test['bio1'] - mean_bio1) / std_bio1 
    env_test['bio17'] = (env_test['bio17'] - mean_bio17) / std_bio17


    file_test = path + f'split_number_{i}/env_testing_sites.csv'


    env_test.to_csv(file_test)
```
